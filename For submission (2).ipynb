```python
MRV_ENABLED = True
class action:
    def __init__(self, input_row, input_column, input_val):
        self.row = input_row
        self.col = input_column
        self.val = input_val

class state:
    def __init__(self, input_board=None):
        if input_board is not None:
            self.board = input_board
        else:
            self.board = []
            for rr in range(9):
                row_list = []
                for cc in range(9):
                    row_list.append(0)
                self.board.append(row_list)
        self.row_candidates = []
        for i in range(9):
            row_cand = []
            for j in range(9):
                row_cand.append(True)
            self.row_candidates.append(row_cand)
        self.col_candidates = []
        for i in range(9):
            col_cand = []
            for j in range(9):
                col_cand.append(True)
            self.col_candidates.append(col_cand)
        self.box_candidates = []
        for i in range(9):
            box_cand = []
            for j in range(9):
                box_cand.append(True)
            self.box_candidates.append(box_cand)
        self.candidate_init()

    def candidate_init(self):
        for rr in range(9):
            for cc in range(9):
                val = self.board[rr][cc]
                if val != 0:
                    self.number_select(rr, cc, val, True)

    def box_i(self, row, col):
        return (row // 3)*3 + (col // 3)

    def number_select(self, row, col, val, initialize=False):
        self.board[row][col] = val
        i = val - 1
        self.row_candidates[row][i] = False
        self.col_candidates[col][i] = False
        box_i = self.box_i(row, col)
        self.box_candidates[box_i][i] = False

    def number_correction(self, row, col, val):
        self.board[row][col] = 0
        i = val - 1
        self.row_candidates[row][i] = True
        self.col_candidates[col][i] = True
        box_i = self.box_i(row, col)
        self.box_candidates[box_i][i] = True

    def constrain_check(self, act):
        if self.board[act.row][act.col] != 0:
            return False
        i = act.val - 1
        box_i = self.box_i(act.row, act.col)
        return (self.row_candidates[act.row][i] and 
                self.col_candidates[act.col][i] and 
                self.box_candidates[box_i][i])

    def apply_action(self, act):
        self.number_select(act.row, act.col, act.val)

    def revert_action(self, act):
        self.number_correction(act.row, act.col, act.val)

    def find_possible_values(self, rr, cc):
        if self.board[rr][cc] != 0:
            return []
        res = []
        for val in range(1,10):
            i = val - 1
            box_i = self.box_i(rr, cc)
            if (self.row_candidates[rr][i] and
                self.col_candidates[cc][i] and
                self.box_candidates[box_i][i]):
                res.append(val)
        return res

def find_empty_cell(s):
    if not MRV_ENABLED:
        for rr in range(9):
            for cc in range(9):
                if s.board[rr][cc] == 0:
                    return rr, cc
        return None
    else:
        min_candidates = 10
        candidates_cells = []
        for rr in range(9):
            for cc in range(9):
                if s.board[rr][cc] == 0:
                    cands = s.find_possible_values(rr, cc)
                    count = len(cands)
                    if count < min_candidates:
                        min_candidates = count
                        candidates_cells = [(rr, cc)]
                    elif count == min_candidates:
                        candidates_cells.append((rr, cc))
        if candidates_cells:
            return candidates_cells[0]
        return None

def solve_sudoku(s, step_count=0):
    empty_cell = find_empty_cell(s)
    if empty_cell is None:
        return True, step_count
    r, c = empty_cell
    candidates = s.find_possible_values(r, c)
    for val in candidates:
        act = action(r, c, val)
        if s.constrain_check(act):
            s.apply_action(act)
            solved, final_step_count = solve_sudoku(s, step_count+1)
            if solved:
                return True, final_step_count
            s.revert_action(act)
    return False, step_count
```

```python
#Q-Learning Functions

class action:
    def __init__(self, row, col, val):
        self.row = row
        self.col = col
        self.val = val

class state:
    def __init__(self, input_board=None):
        if input_board is not None:
            self.board = np.array(input_board, dtype=int)
        else:
            self.board = np.zeros((9,9), dtype=int)
        self.row_candidates = [[True]*9 for _ in range(9)]
        self.col_candidates = [[True]*9 for _ in range(9)]
        self.box_candidates = [[True]*9 for _ in range(9)]
        self.candidate_init()

    def candidate_init(self):
        for r in range(9):
            for c in range(9):
                val = self.board[r][c]
                if val != 0:
                    self.number_select(r, c, val, True)

    def box_i(self, row, col):
        return (row // 3)*3 + (col // 3)

    def number_select(self, row, col, val, initialize=False):
        self.board[row][col] = val
        i = val - 1
        self.row_candidates[row][i] = False
        self.col_candidates[col][i] = False
        box_i = self.box_i(row, col)
        self.box_candidates[box_i][i] = False

    def number_correction(self, row, col, val):
        self.board[row][col] = 0
        i = val - 1
        self.row_candidates[row][i] = True
        self.col_candidates[col][i] = True
        box_i = self.box_i(row, col)
        self.box_candidates[box_i][i] = True

    def constrain_check(self, act):
        if self.board[act.row][act.col] != 0:
            return False
        i = act.val - 1
        box_i = self.box_i(act.row, act.col)
        return (self.row_candidates[act.row][i] and 
                self.col_candidates[act.col][i] and 
                self.box_candidates[box_i][i])

    def apply_action(self, act):
        self.number_select(act.row, act.col, act.val)

    def revert_action(self, act):
        self.number_correction(act.row, act.col, act.val)

    def find_possible_values(self, row, col):
        if self.board[row][col] != 0:
            return []
        possible_values_list = []
        for val in range(1, 10):
            i = val - 1
            box_i = self.box_i(row, col)
            if (self.row_candidates[row][i] and
                self.col_candidates[col][i] and
                self.box_candidates[box_i][i]):
                possible_values_list.append(val)
        return possible_values_list

    def is_solved(self):
        return np.all(self.board != 0)

def get_possible_actions(s):
    acts = []
    for r in range(9):
        for c in range(9):
            if s.board[r][c] == 0:
                candidates = s.find_possible_values(r, c)
                for v in candidates:
                    acts.append(action(r, c, v))
    return acts

def step(s, act):
    if s.constrain_check(act):
        s.apply_action(act)
        if s.is_solved():
            return s, 50.0, True
        else:
            return s, 0.1, False
    else:
        return s, 0.0, True

def encode_state(s):
    return tuple(s.board.flatten())

Q = {}
alpha = 0.1
gamma = 0.95
epsilon = 1.0
epsilon_decay = 0.995
epsilon_min = 0.01

def get_Q(state_key, act_key):
    return Q.get((state_key, act_key), 0.0)

def update_Q(old_s, a, reward, new_s, done):
    old_sk = encode_state(old_s)
    ak = (a.row,a.col,a.val)
    old_q = get_Q(old_sk, ak)
    if done:
        target = reward
    else:
        new_acts = get_possible_actions(new_s)
        if not new_acts:
            target = reward
        else:
            new_sk = encode_state(new_s)
            next_qs = [get_Q(new_sk,(na.row,na.col,na.val)) for na in new_acts]
            target = reward + gamma*max(next_qs)
    new_q = old_q + alpha*(target - old_q)
    Q[(old_sk, ak)] = new_q

def get_mrv_cell(s):
    min_candidates = 10
    candidates_cells = []
    for r in range(9):
        for c in range(9):
            if s.board[r][c] == 0:
                possible_values = s.find_possible_values(r, c)
                num_candidates = len(possible_values)
                if num_candidates > 0 and num_candidates < min_candidates:
                    min_candidates = num_candidates
                    candidates_cells = [(r, c)]
                elif num_candidates == min_candidates:
                    candidates_cells.append((r, c))
    if candidates_cells:
        return random.choice(candidates_cells)
    else:
        return None

def choose_action(s):
    global epsilon
    acts = get_possible_actions(s)
    if not acts:
        return None

    if MRV_ENABLED:
        cell = get_mrv_cell(s)
        if cell is not None:
            acts = [a for a in acts if a.row == cell[0] and a.col == cell[1]]

    if random.random() < epsilon:
        return random.choice(acts)
    else:
        sk = encode_state(s)
        q_vals = [(get_Q(sk,(a.row,a.col,a.val)), a) for a in acts]
        max_q = max(q_vals, key=lambda x:x[0])[0]
        max_actions = [a for q,a in q_vals if q == max_q]
        return random.choice(max_actions)

def solve_with_Q(s, Q):
    step_count = 0
    max_steps = 10000
    while not s.is_solved() and step_count < max_steps:
        acts = get_possible_actions(s)
        if not acts:
            break
        sk = encode_state(s)
        q_vals = []
        for a in acts:
            q_val = Q.get((sk, (a.row,a.col,a.val)), 0.0)
            q_vals.append((q_val, a))
        max_q = max(q_vals, key=lambda x:x[0])[0]
        best_actions = [a for q,a in q_vals if q == max_q]
        chosen_action = best_actions[0]
        if s.constrain_check(chosen_action):
            s.apply_action(chosen_action)
        else:
            break
        step_count += 1

    return s.is_solved(), s, step_count
```
